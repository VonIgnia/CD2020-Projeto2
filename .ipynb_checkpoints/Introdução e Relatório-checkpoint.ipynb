## Introdução

Através de algoritmos de Machine Learning da biblioteca SciKitLearn, temos como objetivo elaborar uma análise preditiva em que a partir de determinadas características de um veículo, seja possível prever seu valor de venda, ou em alguns casos re-venda, essas características foram melhor exploradas e classificadas [nesse notebook]().

## Relatório
Para realizar essa predição:  

1. Os dados foram separados em dois tipos de datasets diferentes: 
    - No primeiro não há discriminação por nome de veículo, ou seja: na modelagem para determinar o preço de um veículo, os fatores considerados são o fabricante, tipo de carro (sport, sedan, hatch, entre outros), local de origem (A base de dados original é brasileira e a localização está divida por unidades federativas), dessa forma a predição de, por exemplo, o valor de um VOLKSWAGEN SAVEIRO e o valor de um FORD ECOSPORT são feitas em conjunto.  
    - No outro há a discriminação por nome de veículo, ou seja: na modelagem para determinar o preço de um veículo, o modelo é feito exclusivamente com um veículo, os fatores considerados são os mesmos, no entanto a predição do valor de um veículo é feita isoladamente em relação aos outros
   
2. Os dados de cada conjunto de dados foram separados novamente em duas partes:
    - Uma para treinar o modelo a elaborar e fazer predições.  
        - inputs de treinamento
        - resultados de treinamento
    - Outra para verificar a acurácia e qualidade das predições feitas pelo modelo
        - inputs de teste
        - resultados de teste
        
    Essas duas etapas de separação podem ser vistas nas células 5 até 9 do notebook (Seção entitulada "Dados") desse[link](functions_notebook.ipynb), sendo que a quinta célula importa o conjunto de dados original, as sexta e sétima células fazem a separação global(sem discriminação por nome de veículo) e as oitava e nona células fazem a separação individual(com discriminação por nome de veículo)
        
3. Após o processamento feito nas etapas anteriores, o dataset é utilizado para treinar o modelo de Machine Learning da biblioteca SciKitLearn escolhido.
    Escolhemos utilizar os três seguintes modelos.
    Esses modelos são melhor explicados e aprofundados nos seguintes notebooks:
    - Regressão linear      [clique aqui](SKLearn%20Linear%20Regression.ipynb#).
    - Árvore de regressão   [clique aqui](Regression%20Tree.ipynb#).
    - Floresta de regressão [clique aqui](Random%20Forest.ipynb#).  
    
    Esses modelos foram treinados nas células 16 (para os dataframes globais) e 10 até 15 (para os dataframes individuais), na seção entitulada "Pré Processamento de Modelos de Regressão e Classificação do SKLearn".

#### Conclusão para Dados "globais":
Pelo que pode ser observado em várias "rodagens" do código, o modelo mais apropriado é o modelo de Random Forest Regression, pelos seguintes motivos:

- O valor de r_squared ($r^{_2}$), também conhecido como coeficiente de determinação, desse último tende a ser (dependendo da "seed" de treinamento pode não ser) mais próximo de 1 (um), melhor explicando a variância do modelo. <font color="#2A2DD3">  
    Explicação completa em: `https://en.wikipedia.org/wiki/Coefficient_of_determination` </font>
      
      
- Tende a possuir um menor erro quadrático médio (Mean Squared Error), ou seja a média do quadrado da distância ou diferença entre os valores estimados e os valores reais é menor nesse modelo do que nos outros dois.  
    <font color="#2A2DD3"> Explicação completa em:`https://en.wikipedia.org/wiki/Mean_squared_error`</font>
    
    
- Tende a possuir um menor erro absoluto médio (Mean Absolute Error), distância média entre a previsão e o valor real.  
   <font color="#2A2DD3"> Explicação completa em:`https://en.wikipedia.org/wiki/Mean_absolute_error`</font>
