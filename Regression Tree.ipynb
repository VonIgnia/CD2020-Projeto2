### Explicação do Modelo:

#### $1$. Introdução ao Modelo:
Uma árvore de decisões (ou decision tree), é uma ferramenta de suporte de decisões, que utiliza um modelo de decisões tree-like (estrutura similar á de uma árvore, a partir de um "caule", surgem várias ramificações ou "galhos", que por si "hospedam" outras ramificações ou "galhos")que leva em conta as decisões feitas e as implicações e consequências dessas decisões. Essa é uma maneira de exibir um algoritmo que contém apenas declarações de controle condicional.  
Árvores de decisões são comumente utipizadas em pesquisas de opreações, especificamente in análise de decisões, para ajudar a identificar uma estratégia mais provável de atinjir uma meta, no entanto também é uma ferramenta popular para Machine Learning

#### $2$. O que é um modelo de Regressão:   
Modelos de regressão são algoritmos de Machine Learning* baseados em aprendizado supervisionado e são utilizados para performar, realizar, tarefas de Regresão. O valor previsto por um modelo de regressão baseia-se em variáveis independentes, e é usado para encontrar a relação entre as variáveis e projeções. Então determinados dados de variáveis preditivas (características do objeto de estudo em questão que são/podem ser consideradas relevantes quando se tenta estimar o valor de outra caracteristica do objeto de estudo), essas variáveis representam os inputs, os xises de uma função, e uma varíavel de resposta contínua, que representa o output, o ípsilon da função, constroem um modelo preditivo que é capaz de:
- Prever o valor da resposta, o ípsilon, a partir das variáveis preditivas, os xises.
- Entender a relação entre os xises e ípsilon

#### $3$. Funcionamento do modelo:
O modelo funciona de acordo com as etapas abaixo:  

- Considere o Conjunto de Dados, que chamaremos de $D$, com um número $a$ de linhas e um número $n$ de características
- Escolha uma amostra das linhas e algumas das características (preferencialmente as relevantes no que está sendo estudado, já descartando as demais) do conjunto de dados original $D$.
- Essas amostras (as quais foi atribuido ao longo do projeto o nome seed) são 'entregues' à uma Árvore de decisões, isso é a seed "germina" construindo e treinando-a, encontrando um resultado preditivo para árvore.
